```python
import pandas as pd
```

```python
df_titles = pd.read_csv('titles.csv')
```

```python
df_titles['index'] = df_titles.index
```

```python
df_links = pd.read_json('links.json')
```

```python
links = {}
```

```python
for row, value in df_links.iterrows():
    
    if value.source in links:
        links[value.source]["target"].append({'target': value.target, 'weight': value.weight})
    else:
        links[value.source] = {
            "targets": {
                
            },
        }
```

```python
df_titles['links'] = df_titles['index'].apply(lambda x: links[x] if x in links else None)
```
